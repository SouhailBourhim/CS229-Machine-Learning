# 🧮 Linear Regression from Scratch — CS229 Style

## Stanford CS229: Machine Learning (Lecture 2)

In this notebook, we implement linear regression **from scratch** using only NumPy.

We will implement:
1. Feature Normalization
2. Cost Function $J(\theta)$
3. Batch Gradient Descent
4. Stochastic Gradient Descent (SGD)
5. Normal Equation (Closed-form solution)
6. Predictions & Visualization

### Mathematical Foundation

**Hypothesis:** $h_\theta(x) = \theta^T x$

**Cost Function:** $J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2$

**Normal Equation:** $\theta = (X^T X)^{-1} X^T \vec{y}$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
plt.style.use('seaborn-v0_8-whitegrid')
np.random.seed(42)
print("✅ Libraries loaded successfully!")
```

## 1. Create the Dataset

```python
# Features: [size (sq ft), bedrooms, age (years)]
X_raw = np.array([
    [2104, 3, 20], [1416, 2, 15], [1534, 3, 10], [852,  2, 35],
    [1940, 4, 5],  [1755, 3, 12], [1268, 2, 25], [1900, 4, 8],
    [1022, 2, 30], [1736, 3, 18], [1635, 3, 22], [1594, 2, 14]
])
y = np.array([460, 232, 315, 178, 420, 390, 270, 410, 195, 380, 350, 310])
m = len(y)
n = X_raw.shape[1]

df = pd.DataFrame(X_raw, columns=['Size (sqft)', 'Bedrooms', 'Age (years)'])
df['Price ($k)'] = y
print(f"Dataset: {m} examples, {n} features\n")
df
```

## 2. Visualize the Data

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))

axes[0].scatter(X_raw[:, 0], y, color='steelblue', s=80, edgecolors='k', alpha=0.7)
axes[0].set_xlabel('Size (sq ft)'); axes[0].set_ylabel('Price ($k)')
axes[0].set_title('Size vs Price', fontweight='bold')

axes[1].scatter(X_raw[:, 1], y, color='coral', s=80, edgecolors='k', alpha=0.7)
axes[1].set_xlabel('Bedrooms'); axes[1].set_ylabel('Price ($k)')
axes[1].set_title('Bedrooms vs Price', fontweight='bold')

axes[2].scatter(X_raw[:, 2], y, color='seagreen', s=80, edgecolors='k', alpha=0.7)
axes[2].set_xlabel('Age (years)'); axes[2].set_ylabel('Price ($k)')
axes[2].set_title('Age vs Price', fontweight='bold')

plt.tight_layout()
plt.show()
```

## 3. Feature Normalization

$$x_j^{(i)} = \frac{x_j^{(i)} - \mu_j}{\sigma_j}$$

```python
def feature_normalize(X):
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    X_norm = (X - mu) / sigma
    return X_norm, mu, sigma

X_norm, mu, sigma = feature_normalize(X_raw)
X = np.c_[np.ones(m), X_norm]
print(f"X shape: {X.shape}")
```

## 4. Core Functions

```python
def hypothesis(X, theta):
    return X @ theta

def compute_cost(X, y, theta):
    m = len(y)
    errors = hypothesis(X, theta) - y
    return (1 / (2 * m)) * np.sum(errors ** 2)

initial_theta = np.zeros(n + 1)
print(f"Initial cost: {compute_cost(X, y, initial_theta):.2f}")
```

## 5. Batch Gradient Descent

$$\theta := \theta - \frac{\alpha}{m} X^T (X\theta - \vec{y})$$

```python
def batch_gradient_descent(X, y, theta, alpha, num_iterations):
    m = len(y)
    cost_history = []
    for i in range(num_iterations):
        gradient = (1/m) * (X.T @ (hypothesis(X, theta) - y))
        theta = theta - alpha * gradient
        cost_history.append(compute_cost(X, y, theta))
    return theta, cost_history

theta_bgd, cost_history_bgd = batch_gradient_descent(
    X, y, np.zeros(n + 1), alpha=0.1, num_iterations=500
)
print(f"BGD θ: {np.round(theta_bgd, 2)}")
print(f"BGD Final cost: {cost_history_bgd[-1]:.4f}")
```

## 6. Stochastic Gradient Descent

```python
def stochastic_gradient_descent(X, y, theta, alpha, num_epochs):
    m = len(y)
    cost_history = []
    for epoch in range(num_epochs):
        indices = np.random.permutation(m)
        for i in indices:
            xi = X[i:i+1]
            yi = y[i:i+1]
            error = hypothesis(xi, theta) - yi
            theta = theta - alpha * (xi.T @ error).flatten()
        cost_history.append(compute_cost(X, y, theta))
    return theta, cost_history

theta_sgd, cost_history_sgd = stochastic_gradient_descent(
    X, y, np.zeros(n + 1), alpha=0.01, num_epochs=100
)
print(f"SGD θ: {np.round(theta_sgd, 2)}")
print(f"SGD Final cost: {cost_history_sgd[-1]:.4f}")
```

## 7. Normal Equation

$$\theta = (X^T X)^{-1} X^T \vec{y}$$

```python
def normal_equation(X, y):
    return np.linalg.inv(X.T @ X) @ X.T @ y

theta_ne = normal_equation(X, y)
print(f"Normal Eq θ: {np.round(theta_ne, 2)}")
print(f"Normal Eq cost: {compute_cost(X, y, theta_ne):.4f}")
```

## 8. Convergence Comparison

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(cost_history_bgd, 'b-', linewidth=2, label='Batch GD')
axes[0].axhline(y=compute_cost(X, y, theta_ne), color='r', linestyle='--', label='Optimal')
axes[0].set_xlabel('Iteration'); axes[0].set_ylabel('Cost J(θ)')
axes[0].set_title('Batch GD Convergence', fontweight='bold'); axes[0].legend()

axes[1].plot(cost_history_sgd, 'r-', linewidth=2, label='SGD')
axes[1].axhline(y=compute_cost(X, y, theta_ne), color='b', linestyle='--', label='Optimal')
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Cost J(θ)')
axes[1].set_title('SGD Convergence', fontweight='bold'); axes[1].legend()

plt.tight_layout()
plt.show()
```

## 9. Predictions

```python
def predict(features, theta, mu, sigma):
    features_norm = (features - mu) / sigma
    x = np.concatenate([[1], features_norm])
    return x @ theta

test_houses = [[1650, 3, 10], [2400, 4, 5], [900, 2, 40]]
labels = ["Medium 10yr", "Large 5yr", "Small 40yr"]

print("PREDICTIONS")
print("=" * 55)
for house, label in zip(test_houses, labels):
    p = predict(np.array(house), theta_ne, mu, sigma)
    print(f"  {label:<15} → ${p:.1f}k")
```

## Summary

| Method | Iterations | Learning Rate | Exact? |
|--------|-----------|---------------|--------|
| Batch GD | Many | Must choose α | No |
| SGD | Many | Must choose α | No |
| Normal Eq | None | Not needed | Yes |
